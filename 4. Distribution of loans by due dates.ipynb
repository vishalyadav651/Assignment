```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Portfolio_data.csv file
portfolio_data = pd.read_csv('Portfolio_data.csv')

# Group data by due date column and get count of loans for each due date
due_date_count = portfolio_data.groupby('Due Date')['Loan Number'].count().reset_index(name='count')

# Plot line chart to visualize the distribution
plt.figure(figsize=(15, 5))
plt.plot(due_date_count['Due Date'], due_date_count['count'])
plt.title('Distribution of Loans by Due Dates')
plt.xticks(rotation=90)
plt.xlabel('Due Date')
plt.ylabel('Loan Count')

plt.show()
```
