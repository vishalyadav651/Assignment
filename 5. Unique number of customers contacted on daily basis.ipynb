```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CommunicationHistory.csv file
communication_history = pd.read_csv('CommunicationHistory.csv')

# Filter for only completed calls
completed_calls = communication_history[communication_history['Status'] == 'completed']

# Convert call time to datetime object and extract date only
completed_calls['call_date'] = pd.to_datetime(completed_calls['Call time']).dt.date

# Group data by call date and count unique number of customers for each date
unique_customers_per_day = completed_calls.groupby('call_date')['Loan Number'].nunique().reset_index(name='count')

# Plot the resulting data as a line chart
plt.plot(unique_customers_per_day['call_date'], unique_customers_per_day['count'])
plt.title('Unique Number of Customers Contacted per Day')
plt.xlabel('Date')
plt.ylabel('Customer Count')
plt.xticks(rotation=90)
plt.show()
```
