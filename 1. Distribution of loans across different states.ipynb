```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the portfolio data file into a pandas dataframe
portfolio_df = pd.read_csv('Portfolio_data.csv')
```

```python
# Group the data by state and count the number of loans in each state
state_counts = portfolio_df.groupby('State')['Loan Number'].count()
```

```python
# Visualize the distribution using a bar chart
plt.figure(figsize=(15, 3))
plt.bar(state_counts.index, state_counts.values, width=0.5,align='center')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.ylabel('Number of loans')
plt.title('Distribution of loans across different states')
plt.show()
```
