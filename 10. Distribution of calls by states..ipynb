```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Portfolio_data.csv file
portfolio_data = pd.read_csv('Portfolio_data.csv')

# Read CommunicationHistory.csv file
communication_history = pd.read_csv('CommunicationHistory.csv')

# Merge data from both files on customer ID column
merged_data = pd.merge(portfolio_data, communication_history, on='Loan Number')

# Group data by state column and count number of calls for each state
calls_per_state = merged_data.groupby('State')['Call time'].count().reset_index(name='count')

# Plot bar chart to visualize the distribution
plt.figure(figsize=(20, 6))
plt.bar(calls_per_state['State'], calls_per_state['count'])
plt.title('Distribution of Calls by States')
plt.xlabel('State')
plt.ylabel('Call Count')
plt.xticks(rotation=90)
plt.show()
```
