```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the portfolio data file into a pandas dataframe
portfolio_df = pd.read_csv('Portfolio_data.csv')
```

```python
# Group the data by loan amount and count the number of loans for each amount
amount_counts = portfolio_df.groupby('Loan Amount')['Loan Number'].count()
```

```python
# Visualize the distribution using a histogram
sns.histplot(data=portfolio_df, x='Loan Amount', bins=10)
plt.xlabel('Loan amount')
plt.ylabel('Number of loans')
plt.title('Distribution of loans across different amounts')
plt.show()
```

```python
# Visualize the distribution using a density plot
sns.kdeplot(data=portfolio_df, x='Loan Amount')
plt.xlabel('Loan amount')
plt.ylabel('Density')
plt.title('Distribution of loans across different amounts')
plt.show()
```
